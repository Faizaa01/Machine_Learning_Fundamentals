```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression, SGDRegressor, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, mean_absolute_error
```

```python
hours = np.array([2,4,6,8,10]).reshape(-1,1)
marks = np.array([55,60,70,80,88])

model = LinearRegression()
model.fit(hours, marks)

print("m: ", model.coef_[0])
print("c: ", model.intercept_)
```

```python

plt.scatter(hours, marks, color='blue', label="Actual Data")
pred_marks = model.predict(hours)
plt.plot(hours, model.predict(hours), color='red', label="Best Fit Line")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.show()
```

```python


new_hours = np.array([[9]])
prediction = model.predict(new_hours)
print("Predicted Marks: ", prediction[0])
```

```python
pred_marks = model.predict(hours)

mae = mean_absolute_error(marks, pred_marks)
mse = mean_squared_error(marks, pred_marks)
rmse = np.sqrt(mse)
r2 = r2_score(marks, pred_marks)

print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("RMSE: ", rmse)
print("R2 Score: ", r2)
```

```python
sgd = SGDRegressor(
    max_iter = 1000,
    learning_rate = "invscaling",
    eta0 = 0.01,
    random_state = 42
)
sgd.fit(hours, marks)

print("m: ", sgd.coef_[0])
print("c: ", sgd.intercept_)
```

```python
pred_sgd = sgd.predict(hours)
mse_sgd = mean_squared_error(marks, pred_sgd)
r2_sgd = r2_score(marks, pred_sgd)

print("Mean Squared Error: ", mse_sgd)
print("R2 Score: ", r2_sgd)
```

```python
plt.scatter(hours, marks, color='blue', label="Actual Data")
pred_marks = model.predict(hours)
plt.plot(hours, pred_marks, label="LinearRegression", linestyle="--")
plt.plot(hours, pred_sgd, label="SGDRegression", linestyle="-.")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.show()
```
