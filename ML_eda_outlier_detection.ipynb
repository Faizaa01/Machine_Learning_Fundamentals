```python
import numpy as np
import pandas as pd
```

```python
salaries = np.array([45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500])

range = salaries.max() - salaries.min()
range
```

```python
varience = salaries.var()
std_dev = salaries.std()
varience , std_dev
```

```python
Q1 = np.percentile(salaries, 25)
Q2 = np.percentile(salaries, 50)
Q3 = np.percentile(salaries, 75)
Q1, Q2, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
Lower_fence = Q1 - 1.5 * IQR
Upper_fence = Q3 + 1.5 * IQR
Lower_fence, Upper_fence
```

```python
Outliers = salaries[(salaries < Lower_fence) | (salaries > Upper_fence)]
Outliers
```
