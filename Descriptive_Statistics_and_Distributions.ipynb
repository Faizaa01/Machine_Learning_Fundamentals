```python
Cat_weights = np.array([2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5])
P25 = np.percentile(Cat_weights, 25)
P50 = np.percentile(Cat_weights, 50)
P75 = np.percentile(Cat_weights, 75)

P25, P50, P75
```

```python
IQR = P75 - P25
lower = P25 - 1.5 * IQR
upper = P75 + 1.5 * IQR

Outliers = Cat_weights[(Cat_weights<lower) | (Cat_weights>upper)]
Outliers
```

```python
# Confusion matrix
TP = 120
FN = 30
FP = 80
TN = 770

Total = TP + TN + FP + FN

Accuracy = (TP + TN) / Total
Precision = TP / (TP + FP)
Recall = TP / (TP + FN)
F1 = 2 * Precision * Recall / (Precision + Recall)
Specificity = TN / (TN + FP)
NPV = TN / (TN + FN)
Prevalence = (TP + FN) / Total

Accuracy, Precision, Recall, F1, Specificity, NPV, Prevalence
```

```python
Weight_kg = np.array([58, 62, 65, 66, 190], dtype=float)
Height_cm = np.array([150, 160, 170, 175, 180], dtype=float)

def standardize_first_three(x):
    x_subset = x[:3]
    mean = np.mean(x_subset)
    std = np.std(x_subset)
    z_scores = (x_subset - mean) / std
    return z_scores

z_weight = standardize_first_three(Weight_kg)
z_height = standardize_first_three(Height_cm)

print("Standardization (first 3 values):")
print("Weight:", z_weight)
print("Height:", z_height)
```

```python
def min_max_scaling(x):
    return (x - x.min()) / (x.max() - x.min())

weight_minmax = min_max_scaling(Weight_kg)
height_minmax = min_max_scaling(Height_cm)

print("\nMin-Max Scaling [0,1]:")
print("Weight:", weight_minmax)
print("Height:", height_minmax)
```

```python
def robust_scaling(x):
    median = np.median(x)
    Q1 = np.percentile(x, 25)
    Q3 = np.percentile(x, 75)
    IQR = Q3 - Q1
    return (x - median) / IQR

weight_robust = robust_scaling(Weight_kg)
print("\nRobust Scaling (Weight):", weight_robust)
```

```python
import numpy as np

a = np.array([3, -1, 2])
b = np.array([4, 0, -2])

dot_ab = np.dot(a, b)

l1_a = np.linalg.norm(a, ord=1)
l2_a = np.linalg.norm(a)

print("Dot product a·b:", dot_ab)
print("L1 norm of a:", l1_a)
print("L2 norm of a:", l2_a)
```
